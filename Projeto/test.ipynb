```julia
N = 100  # Número de vértices
p = 0.5  # Probabilidade de criação de uma aresta
L = 40
```

```julia
rede = criar_rede_erdos_renyi(N, p, 42)
x_AG,y_AG = AG(rede, N, L, 42, 300, 6000,0.01)[1:2]
x_SA,y_SA = SimulatedAnneling(rede, N, L, 42, 1, 21,5000,0.5)[1:2]
x_LS,y_LS = LocalSearch(rede, N, L, 42, 2, 17)[1:2]
```
